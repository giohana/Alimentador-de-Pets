- Carregamento e pré processamento de dados

```
import tensorflow as tf
import tensorflow_datasets as tfds

# Carregar dataset (exemplo com Food-101)
dataset, info = tfds.load("food101", as_supervised=True, with_info=True)
train_data, test_data = dataset["train"], dataset["validation"]

# Normalização
def preprocess(image, label):
    image = tf.image.resize(image, (224, 224)) / 255.0
    return image, label

train_data = train_data.map(preprocess).batch(32).shuffle(1000)
test_data = test_data.map(preprocess).batch(32)
```
